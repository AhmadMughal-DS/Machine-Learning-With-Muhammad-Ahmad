```python
import numpy as np

# Calculate the mean of the 'Age' column
mean_age = df['Age'].mean()

# Calculate the third quartile (Q3) of the 'Age' column
q3_age = np.quantile(df['Age'].dropna(), 0.75)

# Fill missing values with the sum of mean and Q3
new_age = df['Age'].fillna(mean_age + q3_age)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
df['Age'].hist(bins=50,ax=ax,density=True,color='red')
new_age.hist(bins=50,ax=ax,density=True,color='green',alpha=0.8)
```

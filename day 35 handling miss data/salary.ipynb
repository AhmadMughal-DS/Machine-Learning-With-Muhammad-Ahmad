```python
new = df.dropna()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df['Age'].hist(bins=50,ax=ax,density=True,color = 'red')
new['Age'].hist(bins=50,ax=ax,density=True,color = 'green',alpha = 0.7)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df['Years of Experience'].hist(bins=50,ax=ax,density=True,color = 'red')
new['Years of Experience'].hist(bins=50,ax=ax,density=True,color = 'green',alpha = 0.7)

```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df['Salary'].hist(bins=50,ax=ax,density=True,color = 'red')
new['Salary'].hist(bins=50,ax=ax,density=True,color = 'green',alpha = 0.7)
```

```python
temp = pd.concat([
    df['Gender'].value_counts()/len(df),
    new['Gender'].value_counts()/len(df)
],axis=1)
temp.columns = ['original','cca']
temp
```

```python
temp = pd.concat([
    df['Education Level'].value_counts()/len(df),
    new['Education Level'].value_counts()/len(df)
],axis=1)
temp.columns = ['original','cca']
temp
```

```python
temp = pd.concat([
    df['Job Title'].value_counts()/len(df),
    new['Job Title'].value_counts()/len(df)
],axis=1)
temp.columns = ['original','cca']
temp
```
